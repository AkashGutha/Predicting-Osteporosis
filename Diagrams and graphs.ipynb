```python
# import matplotlib for graphs
import  matplotlib
from matplotlib import pyplot

# set matplotlib to work inline
%matplotlib inline
```

```python
# import data file
import pandas as pd

data_unscaled = pd.read_csv('data_processed_unscaled.csv')
data_scaled = pd.read_csv('data_processed_scaled.csv')
```

```python
averages_unscaled = data_unscaled['AVG']
averages_scaled = data_scaled['AVG']
```

```python
# frequency by age
pyplot.scatter(averages_unscaled, data_unscaled['AGE'])
pyplot.show()
```

```python
pyplot.scatter(averages_scaled, data_scaled['AGE'])
pyplot.show()
```
